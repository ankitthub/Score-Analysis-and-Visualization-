```python
#drop unnamed column
df=df.drop("Unnamed: 0", axis=1)
print(df.head())
```

```python
df.head()
```

```python
#gender 
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
#from above chart female is more in number
```

```python
gb= df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gp)
```

```python
#drawing the heatmap
plt.figure(figsize=(5,5))
sns.heatmap(gb, annot=True)
plt.title("Relationship btween Parent Education and student score")
plt.show()
```

```python
gb1= df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1, annot=True)
plt.title("Relationship btween Parent Marital Status and student score")

plt.show()
```

```python
#from the above figure we can conclude that education of the parents is having the good impact on students mark
#no/negligible impact of marital status of parents on students mark
```
